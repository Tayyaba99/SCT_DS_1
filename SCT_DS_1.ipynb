```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Population of India.csv',  encoding='utf-8')
df.head()
```

```python
df_final = df.drop(['State/UT', 'Population[50]'], axis=1)
```

```python
# Assuming 'State/UT' and 'Population[50]' are column names in your data
plt.figure(figsize=(10, 5))
plt.bar(df['State/UT'], df['Population[50]'])
plt.title('Population by State/UT')
plt.xlabel('State/UT')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

```python
# Assuming 'Population[50]' is a numerical column
plt.figure(figsize=(10, 5))
plt.hist(df['Population[50]'], bins=20, color='green', edgecolor='black')
plt.title('Population Distribution')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Assuming the latest year column is the last one in the dataframe
latest_year = df.columns[-1]

plt.figure(figsize=(10, 5))
plt.hist(df[latest_year], bins=20, color='green', edgecolor='black')
plt.title('Population Distribution')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
